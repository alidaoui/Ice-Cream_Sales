# Step 0: Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Step 1: Import Data

```python
df=pd.read_csv("IceCreamData.csv")
```

# Step 3: Create Training and Testing Dataset

```python
X=df[['Temperature']]
y=df[['Revenue']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

# Step 4: Train Model

```python
from sklearn.linear_model import LinearRegression #Import Classifier
regressor = LinearRegression(fit_intercept = True) #fit_intercept determines if model will find intercept value
regressor.fit(X_train, y_train)
```

```python
print('Linear Coefficient (m) is',regressor.coef_)
print('Linear Coefficient (b) is',regressor.intercept_)
```

# Step 5: Test Model

```python
y_predict=regressor.predict(X_test)
```

```python
#Visualize Results
plt.figure(figsize=(7,7))
plt.scatter(x=X_train, y=y_train, color='Gray')
plt.plot(X_train,regressor.predict(X_train),color='Red')
plt.title('Ice Cream Revenue as a Function of Temperature (Training set)')
plt.ylabel('Revenue(in USD)')
plt.xlabel('Temperature(in degrees celsius)')
```

```python
#Visualize Results
plt.figure(figsize=(7,7))
plt.scatter(x=X_test, y=y_test, color='Gray')
plt.plot(X_test,regressor.predict(X_test),color='Blue')
plt.title('Ice Cream Revenue as a Function of Temperature (Testing set)')
plt.ylabel('Revenue(in USD)')
plt.xlabel('Temperature(in degrees celsius)')
```
